**กรณีศึกษาที่ 1: ปักหมุดขั้นพื้นฐาน  Create Marker ปักหมุดลงบนแผนที่**

พิมพ์หรือคัดลอก code ชุดนี้ลงใน google colab แล้ว ทดลอง run แล้วดูผลลัพธ์
จากผลลัพธ์ที่ได้ นักเรียนเห็นอะไรบ้าง

```python
import folium #เรียกใช้ folium library
lat, lon = 13.7649, 100.5383 #กำหนดตัวแปร ละติจูด และ ลองจิจูด
m = folium.Map(location=[lat, lon], zoom_start=17) #สั่งสร้างแผนที่
folium.Marker( #คำสั่งสำหรับสร้างหมุด
    [13.7649, 100.5383], #พิกัดที่ต้องการปักหมุด
    popup="อนุสาวรีย์ชัยสมรภูมิ", #เมื่อคลิกที่หมุดให้แสดงข้อความนี้
    tooltip="คลิกเลย", #เพิ่มเครื่องหมายลูกน้ำ 1 ตัว ในบรรทัดนี้
    # เพิ่มบรรทัดนี้เข้าไปเพื่อเปลี่ยนสีหมุด
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m) #แนบคำสั่งนี้ลงในคำสั่งสร้างแผนที่
m #แสดงผลแผนที่
```

**กรณีศึกษาที่ 2: เรียกใช้ไอคอนพิเศษจาก https://fontawesome.com/**

ทดลองเปลี่ยนไอคอน เป็นรูปแบบอื่นๆ โดยการใช้ไอคอนพิเศษแบบ FontAwesome (แนะนำ เพราะมีเยอะมาก) แต่จะต้องใส่พารามิเตอร์ prefix='fa' เข้าไปในคำสั่งด้วย

```python
import folium #เรียกใช้ folium library
lat, lon = 13.7649, 100.5383 #กำหนดตัวแปร ละติจูด และ ลองจิจูด
m = folium.Map(location=[lat, lon], zoom_start=17) #สั่งสร้างแผนที่
folium.Marker( #คำสั่งสำหรับสร้างหมุด
    [13.7649, 100.5387], #พิกัดที่ต้องการปักหมุด เยื้องไปด้านข้างนิดหน่อย เพิ่มลองจิจูด
    popup="อนุสาวรีย์ชัยสมรภูมิ", #เมื่อคลิกที่หมุดให้แสดงข้อความนี้
    tooltip="คลิกเลย", #เพิ่มเครื่องหมายลูกน้ำ 1 ตัว ในบรรทัดนี้
    # เพิ่มบรรทัดนี้เข้าไปเพื่อเปลี่ยนสีหมุด
    icon=folium.Icon(color='blue', icon='ghost', prefix='fa')
).add_to(m) #แนบคำสั่งนี้ลงในคำสั่งสร้างแผนที่
m #แสดงผลแผนที่
```

**กรณีศึกษาที่ 3: การใช้รูปทรงเลขาคณิตแทนหมุดแผนที่:รูปวงกลม**

```python
import folium
lat, lon = 13.7649, 100.5383
m = folium.Map(location=[lat, lon], zoom_start=17)

folium.CircleMarker(
    location=[13.7649, 100.5383 ],
    radius=10,      # ขนาดรัศมี (หน่วยเป็นพิกเซล)
    color='red',    # สีเส้นขอบ เช่น red, blue, green, purple, หรือ orange เป็นต้น
    fill=True,      # True สั่งให้ระบายสีข้างใน False สั่งให้ไม่ต้องระบายสีข้างใน
    stroke=False,   # True สั่งให้ระบายสีเส้นขอบ False สั่งให้ไม่ต้องระบายสีเส้นขอบ
    fill_color='red', # สีพื้นหลังข้างใน เช่น red, blue, green, purple, หรือ orange เป็นต้น
    fill_opacity=0.7, # ความโปร่งแสง (0.0-1.0)
    popup='จุดสีแดง'
).add_to(m)

m
```

**กรณีศึกษาที่ 4: การใช้รูปทรงเลขาคณิตแทนหมุดแผนที่:รูปหลายเหลี่ยม**

```python
import folium
lat, lon = 13.7649, 100.5383
m = folium.Map(location=[lat, lon], zoom_start=17)

folium.RegularPolygonMarker(
    location=[13.7649, 100.5383 ],

    number_of_sides=3,  # กำหนด 3 เหลี่ยม
    rotation=210,        # หมุน 210 องศา

    radius=10,      # ขนาดรัศมี (หน่วยเป็นพิกเซล)
    color='red',    # สีเส้นขอบ เช่น red, blue, green, purple, หรือ orange เป็นต้น
    fill=True,      # True สั่งให้ระบายสีข้างใน False สั่งให้ไม่ต้องระบายสีข้างใน
    stroke=False,   # True สั่งให้ระบายสีเส้นขอบ False สั่งให้ไม่ต้องระบายสีเส้นขอบ
    fill_color='red', # สีพื้นหลังข้างใน เช่น red, blue, green, purple, หรือ orange เป็นต้น
    fill_opacity=0.7, # ความโปร่งแสง (0.0-1.0)
    popup='สามเหลี่ยม',

    gradient=False, # True ไล่สีอ่อนไปเข้ม False ไม่ต้องไล่สี

).add_to(m)

m
```

**แบบฝึกหัดที่ 1: สร้างแผนที่บอกพิกัดโรงเรียนของนักเรียน พร้อมปักหมุดแนะสถานที่สำคัญรอบๆ ของโรงเรียนแบบเจ้าถิ่น**
