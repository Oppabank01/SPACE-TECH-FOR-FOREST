**กรณีศึกษาที่ 1: พื้นขาวสว่าง (light mode)**

```python
import folium
lat, lon = 13.768038188064454, 100.42676017803777
m = folium.Map(location=[lat, lon], zoom_start=17)

folium.TileLayer('CartoDB positron', name='แผนที่สีขาว').add_to(m) #เพิ่มบรรทัดนี้เพื่อปรับรูปแบบพื้นหลัง

folium.Marker(
    [13.768038188064454, 100.42676017803777],
    popup="โรงเรียนโพธิสารพิทยากร",
    tooltip="คลิกเลย",
    icon=folium.Icon(color='blue', icon='school', prefix='fa')
).add_to(m)
m
```

**กรณีศึกษาที่ 9: การสร้าง "แผงควบคุมการเปิด-ปิดเลเยอร์" บนแผนที่**

```python
import folium
lat, lon = 13.768038188064454, 100.42676017803777
# ตั้งค่า zoom_start เท่ากับ 18
m = folium.Map(location=[lat, lon], zoom_start=18, tiles=None)

# สร้าง Layer แต่ละแบบ
folium.TileLayer('OpenStreetMap', name='แผนที่ถนนปกติ').add_to(m)
folium.TileLayer('CartoDB positron', name='แผนที่สีขาว').add_to(m)
folium.TileLayer('CartoDB dark_matter', name='แผนที่สีดำ').add_to(m)
folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Esri',
    name='ภาพดาวเทียม (Esri)'
).add_to(m)

# สร้าง Marker
folium.Marker(
    [lat, lon],
    popup="โรงเรียนโพธิสารพิทยากร",
    tooltip="คลิกเลย",
    icon=folium.Icon(color='blue', icon='school', prefix='fa')
).add_to(m)

# สร้างปุ่มควบคุม (collapsed=False เพื่อให้เห็นเมนูเลย ไม่ต้องกดเปิด)
folium.LayerControl(collapsed=False).add_to(m)

m
```
