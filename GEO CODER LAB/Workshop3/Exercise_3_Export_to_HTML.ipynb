**กรณีศึกษาที่ 1: การบันทึกแผนที่เป็น HTML ไฟล์**

```python
import folium

lat, lon = 13.768038188064454, 100.42676017803777

# แก้จุดที่ 1: ใส่ tiles=None เพื่อล้างค่าเริ่มต้นออก
# แก้จุดที่ 2: ลด zoom_start เหลือ 18 (เลข 20 สูงไปสำหรับ OSM บางทีภาพจะไม่ขึ้น)
m = folium.Map(location=[lat, lon], zoom_start=18, tiles=None)

# สร้าง Layer แต่ละอัน
folium.TileLayer('OpenStreetMap', name='แผนที่ถนนปกติ').add_to(m)
folium.TileLayer('CartoDB positron', name='แผนที่สีขาว').add_to(m)
folium.TileLayer('CartoDB dark_matter', name='แผนที่สีดำ').add_to(m)

folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Esri',
    name='ภาพดาวเทียม (Esri)'
).add_to(m)

# สร้าง Marker
folium.Marker(
    [lat, lon],
    popup="โรงเรียนโพธิสารพิทยากร",
    tooltip="คลิกเลย",
    icon=folium.Icon(color='blue', icon='school', prefix='fa')
).add_to(m)

# สร้างปุ่มควบคุม (collapsed=False เพื่อให้เห็นเมนูเลย ไม่ต้องกดเปิด)
folium.LayerControl(collapsed=True).add_to(m)

m.save('my_school_location_map.html')

m
```
